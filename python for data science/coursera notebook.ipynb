### The code should generate the fibonacci sequence.

````python
def fibonacci(max=10):
    a = [1,1]

    for i in range(1,max-1):
        #a.append(i)
        next = a[i] + a[i-1]
        #print(i,a[i],a[i-1], next)
        a.append(next)
       # i += 1
 
    return a

#print(fibonacci(30))
import numpy as np
import seaborn as sns

sns.distplot(fibonacci(10))
````
